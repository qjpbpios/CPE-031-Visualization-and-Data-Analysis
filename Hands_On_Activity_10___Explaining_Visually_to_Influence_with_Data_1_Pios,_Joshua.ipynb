**4. Procedures:**

Step 1. Import Libraries and Create Sample Data

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample sales dataset
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Sales": [200, 250, 180, 300, 280, 350],
    "Marketing_Spend": [50, 60, 45, 70, 65, 80]
}
df = pd.DataFrame(data)
df
```

Step 2. Know Your Audience — Compare Chart Styles

```python
# Executive audience: prefer clarity (simple bar)
plt.figure(figsize=(7,4))
sns.barplot(x="Month", y="Sales", data=df, color="steelblue")
plt.title("Executive View: Monthly Sales Overview", fontsize=13)
plt.show()

# Analyst audience: prefer detailed trends (line chart)
plt.figure(figsize=(7,4))
sns.lineplot(x="Month", y="Sales", data=df, marker="o", label="Sales")
sns.lineplot(x="Month", y="Marketing_Spend", data=df, marker="o", label="Marketing Spend")
plt.title("Analyst View: Sales vs Marketing Trend", fontsize=13)
plt.legend()
plt.show()
```

**Task 1:** Create one additional visualization suited for a different audience (e.g., a customer, investor, or teammate).

```python
plt.figure(figsize=(8, 8))
plt.pie(df["Sales"], labels=df["Month"], autopct='%1.1f%%', startangle=140, colors=sns.color_palette("viridis", len(df)))
plt.title("Investor View: Monthly Sales Contribution", fontsize=13)
plt.show()
```

Step 3. Know Your Message — Highlight What Matters

```python
# Highlight the highest sales month (preattentive color cue)
highlight = df["Sales"].idxmax()

colors = ["gray" if i != highlight else "crimson" for i in range(len(df))]
plt.figure(figsize=(7,4))
plt.bar(df["Month"], df["Sales"], color=colors)
plt.title("Highlighting the Best Month in Sales", fontsize=13)
plt.show()
```

**Task 2:** Use color or annotations to emphasize a key insight (e.g., lowest point, sudden increase).

```python
lowest_sales_month = df["Sales"].idxmin()

colors = ["gray" if i != lowest_sales_month else "orange" for i in range(len(df))]
plt.figure(figsize=(7,4))
plt.bar(df["Month"], df["Sales"], color=colors)
plt.title("Highlighting the Lowest Month in Sales", fontsize=13)

plt.annotate(f'Lowest Sales: {df["Sales"][lowest_sales_month]}',
             xy=(lowest_sales_month, df["Sales"][lowest_sales_month]),
             xytext=(lowest_sales_month, df["Sales"][lowest_sales_month] + 30),
             arrowprops=dict(facecolor='black', shrink=0.05),
             ha='center')

plt.show()
```

Step 4. Storytelling with Charts — Beginning, Middle, End

```python
# Beginning: Context
plt.figure(figsize=(7,4))
sns.barplot(x="Month", y="Sales", data=df, color="skyblue")
plt.title("Beginning: Sales Started Steady", fontsize=13)
plt.show()

# Middle: Change or Challenge
plt.figure(figsize=(7,4))
sns.lineplot(x="Month", y="Sales", data=df, color="orange", marker="o")
plt.title("Middle: A Drop in March Due to Low Marketing", fontsize=13)
plt.show()

# End: Resolution
plt.figure(figsize=(7,4))
sns.barplot(x="Month", y="Sales", data=df, color="limegreen")
plt.title("End: Recovery and Growth After March", fontsize=13)
plt.show()
```

**Task 3:** Write one or two lines describing your data story from beginning → middle → end.

*   The data story shows steady initial sales, a drop in March, followed by subsequent recovery and growth.

Step 5. Bringing It All Together — Presentation Design

```python
# Simple combined story chart
plt.figure(figsize=(8,5))
sns.lineplot(x="Month", y="Sales", data=df, marker="o", linewidth=2, color="royalblue", label="Sales")
sns.lineplot(x="Month", y="Marketing_Spend", data=df, marker="o", linestyle="--", color="orange", label="Marketing Spend")

plt.fill_between(df["Month"], df["Sales"], alpha=0.1, color="royalblue")
plt.title("From Effort to Impact: Marketing Spend Drives Sales Growth", fontsize=13)
plt.legend()
plt.show()
```

**5. Supplementary Activity:**

1. Use a small dataset available online (kaggle, google dataset, etc.).

2. Visualize it using a bar chart.

3. Highlight one important data point (e.g., the highest or lowest value) using a distinct color.

4. Add a title and annotation that tell the story behind the data.

```python
import pandas as pd
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv"

try:
    response = requests.get(url)
    response.raise_for_status()
    data = StringIO(response.text)
    df_online = pd.read_csv(data)
    print("Dataset loaded successfully.")
    print(df_online.head())
except requests.exceptions.RequestException as e:
    print(f"Error downloading dataset: {e}")
    print("Please ensure the URL is correct and accessible.")
except Exception as e:
    print(f"An error occurred while loading the dataset: {e}")
```

```python
df_online.info()
print(df_online.head())
```

```python
df_2007 = df_online[df_online["year"] == 2007]
df_top_10_gdp = df_2007.sort_values(by="gdpPercap", ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x="country", y="gdpPercap", data=df_top_10_gdp)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
highest_gdp_country = df_top_10_gdp.loc[df_top_10_gdp['gdpPercap'].idxmax()]
highest_gdp_country_name = highest_gdp_country['country']

colors = ["steelblue" if country == highest_gdp_country_name else "gray" for country in df_top_10_gdp['country']]

plt.figure(figsize=(10, 6))
sns.barplot(x="country", y="gdpPercap", data=df_top_10_gdp, palette=colors)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
highest_gdp_country = df_top_10_gdp.loc[df_top_10_gdp['gdpPercap'].idxmax()]
highest_gdp_country_name = highest_gdp_country['country']
highest_gdp_value = highest_gdp_country['gdpPercap']
highest_gdp_index = df_top_10_gdp[df_top_10_gdp['country'] == highest_gdp_country_name].index[0]
x_coordinate = df_top_10_gdp.index.get_loc(highest_gdp_index)


colors = ["steelblue" if country == highest_gdp_country_name else "gray" for country in df_top_10_gdp['country']]

plt.figure(figsize=(10, 6))
ax = sns.barplot(x="country", y="gdpPercap", data=df_top_10_gdp, palette=colors)
plt.title("Top 10 Countries by GDP per Capita in 2007")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.annotate(f'{highest_gdp_value:.2f}',
             xy=(x_coordinate, highest_gdp_value),
             xytext=(x_coordinate, highest_gdp_value + 2000),
             arrowprops=dict(facecolor='black', shrink=0.05),
             ha='center')

plt.show()
```
