Data Title: School Supplies Purchases

---


Data Interpretation: This dataset shows sales of school supplies, like pencils and binders, with details on when and where they were sold, who sold them, how many were bought, their cost, and the total price. We can use this data to see what sold well, which areas had the most sales, and who the best salespeople were over time.

**Similar Selling Date**

```python
df[df['OrderDate'] == '12/29/21']
```

```python
df_date = df[df['OrderDate'] == '12/29/21']

sales_summary_on_date = df_date.groupby(['Region', 'Rep'])[['Units', 'Unit Cost', 'Total']].sum()

plt.figure(figsize=(8, 6))
sales_summary_on_date['Units'].plot(kind='bar', color='skyblue', edgecolor='black')

plt.title('Total Units Sold by Region and Representative on 12/29/21')
plt.xlabel('Region, Representative')
plt.ylabel('Total Units Sold')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

**Data Interpretation**

The bar chart shows the total units sold by each representative in each region on December 29, 2021. Based on the chart, only one representative, Parent from the East region, had sales on this date, selling a total of 148 units.

---



**By Region**

```python
df[df['Region'] == 'East']
```

```python
df_east_region = df[df['Region'] == 'East']

units_of_rep_in_east = df_east_region.groupby('Rep')['Units'].sum()

plt.figure(figsize=(10, 6))
units_of_rep_in_east.plot(kind='bar', color='skyblue', edgecolor='black')

plt.title('Total Units Sold by Representative in the East Region')
plt.xlabel('Representative')
plt.ylabel('Total Units Sold')
plt.xticks(rotation=45, ha='right')

plt.show()
```

**Data Interpretation**

This bar chart displays the total units sold by each representative in the East region across all dates in the dataset. It clearly shows that Jones sold the highest number of units in the East region (380 units), followed by Parent (244 units), and then Howard (125 units).

---



**By Rep**

```python
df[df['Rep'] == 'Jones']
```

```python
df_jones = df[df['Rep'] == 'Jones']

units_by_item_jones = df_jones.groupby('Item')['Units'].sum()

plt.figure(figsize=(10, 6))
units_by_item_jones.plot(kind='bar', color='skyblue', edgecolor='black')

plt.title('Total Units Sold by Item for Jones')
plt.xlabel('Item')
plt.ylabel('Total Units Sold')
plt.xticks(rotation=45, ha='right')

plt.show()
```

**Data Interpretation**

The item that Jones sold the most, regardless of the region, was Pencil, with a total of 130 units sold. Following Pencils, Jones sold 124 units of Binders, 64 units of Pens, and 62 units of Pen Sets.
