**Preattentive Attributes**

Preattentive attributes are visual features that our brain processes effortlessly and almost instantly—without conscious effort—via iconic memory. Leveraging these attributes in data visualizations helps reduce cognitive load and quickly draw attention to the most important elements.

**Four Key Types of Preattentive Attributes**

1. Color <br>

Color is one of the most powerful preattentive attributes because it stands out immediately in a visual field.

Hue: Refers to the “name” of the color (e.g., red, blue, green). Hue is often used to group categories or highlight distinctions. Example: red dots among gray dots are instantly noticeable.

Saturation: Refers to the intensity of a color. A bright red circle pops out against dull pink ones, even if they share the same hue.

Luminance (Brightness): Refers to how light or dark a color appears. A light yellow square will stand out against dark gray ones.



image.png

2. Form <br>

Form refers to how an object’s visual shape and structure influence perception. Several sub-attributes fall here:

Orientation: Angles or directions can make certain elements stand out (e.g., a diagonal line among horizontal ones).

Size: Bigger objects capture attention first. Example: a large bubble in a bubble chart stands out immediately.

Shape: Different shapes can indicate grouping. Example: using triangles vs circles in a scatter plot to show two categories.

Length: Humans are very good at comparing lengths (bar chart bars), making length a highly effective way to show quantitative differences.

Width: Thicker or thinner elements can communicate weight or importance (e.g., flow widths in Sankey diagrams).

3. Spatial Positioning <br>

Spatial positioning refers to where elements are placed in relation to each other. Humans are naturally skilled at noticing patterns, clustering, and alignment in space.



*   X and Y positions: In scatter plots, our eyes instantly see clusters, trends, or outliers.
*   Alignment: Objects aligned differently from others draw attention.
*   Grouping: Proximity of objects suggests relationships (Gestalt principle of proximity).

4. Movement <br>

Movement is a strong but less commonly used preattentive attribute (especially in static charts). Our brains are highly sensitive to motion—it immediately captures attention.



*   Flicker/Animation: Blinking elements or pulsing dots draw the eye faster than color or size changes.
*   Motion Path: Direction of movement can show flow or change over time.
*   Dynamic Visualization: Used more in digital dashboards and interactive graphics rather than static charts.

```python
# Import libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Step 1: Sample data
np.random.seed(42)
data = pd.DataFrame({
    'X': np.random.randint(1, 100, 30),
    'Y': np.random.randint(1, 100, 30),
    'Category': np.random.choice(['A', 'B', 'C'], 30),
    'Size': np.random.randint(50, 300, 30)
})
```

```python
# Step 2: Map categories to colors and shapes
colors = {'A': 'red', 'B': 'blue', 'C': 'green'}
shapes = {'A': 'o', 'B': 's', 'C': '^'}
```

```python
# Step 3: Plot scatter points for each category
plt.figure(figsize=(10,6))
for cat in data['Category'].unique():
    subset = data[data['Category'] == cat]
    plt.scatter(subset['X'], subset['Y'],
                s=subset['Size'],        # Size of points
                c=colors[cat],           # Color by category
                marker=shapes[cat],      # Shape by category
                alpha=0.7,
                edgecolors='black',
                label=f'Category {cat}')

# Step 4: Add labels, title, legend
plt.title('Preattentive Attributes Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(title='Category')
plt.grid(True)
plt.show()
```

**Gestalt Principles**

refer to the guiding principles of how people interpret and perceive what they see. These principles can be used in the design of effective data visualizations. The principles generally describe how people define order and meaning in things that they see. We will limit our discussion to the four Gestalt principles that are most closely related to the design of data visualizations: similarity, proximity, enclosure, and connection. An understanding of these principles can help in creating more effective data visualizations and help differentiate between clutter and meaningful design in data visualizations.

1. Similarity <br>
The Gestalt principle of similarity suggests that people naturally see objects with shared features such as color, shape, size, orientation, or other preattentive attributes as part of the same group. In data visualization, this means that when elements share these characteristics, the audience will interpret them as being related. In the example below, objects with the same color or shape are perceived as grouped together. Therefore, when designing visualizations, it is essential to use similar attributes only for elements that are meant to be grouped, to avoid misleading the audience.

image.png

2. Proximity <br>
The Gestalt principle of proximity states that people naturally perceive objects that are close together as belonging to the same group, while objects farther apart are seen as separate groups. This principle is often evident in data visualizations such as scatter charts. For example, if a company wants to analyze its customers by age and annual income, a scatter chart of these variables can reveal natural groupings. In the example below, we tend to see three distinct clusters of customers based on how close the points are to each other. This illustrates the Gestalt principle of proximity.



image.png

3. Enclosure <br>
The Gestalt principle of enclosure states that objects that are physically enclosed together are perceived as belonging to the same group. This can be illustrated in two ways. First, we can reinforce the similarity principle by enclosing points that are already close to one another, as shown below. Alternatively, if there is a third attribute of the customers such as educational background that we want to highlight, we can use enclosure to indicate this grouping even when the customers are not close together in the chart. Enclosure can be represented in different ways, such as shaded areas or dashed boxes. In most cases, even a simple suggestion of enclosure is enough for the audience to perceive the objects as part of the same group.

4. Connection <br>

The Gestalt principle of connection suggests that people perceive objects that are linked or connected as part of the same group. A common application of this principle in data visualization is in time-series data. For example, a data center company may want to compare its forecasted server loads with the actual loads experienced by customers over a two-week period. By connecting the data points, it becomes easier to see trends and differences between the forecasted and actual peak server loads.

```python
# Step 1: Create sample dataset
np.random.seed(2025)
data = pd.DataFrame({
    'Age': np.random.randint(18, 60, 30),
    'Income': np.random.randint(30000, 120000, 30),
    'Education': np.random.choice(['High School', 'Bachelor', 'Master'], 30),
    'CustomerType': np.random.choice(['Regular', 'Premium', 'VIP'], 30),
    'VisitFrequency': np.random.randint(1, 15, 30)
})
```

```python
# Step 2: Map customer type to colors and shapes (Similarity)
colors = {'Regular': 'red', 'Premium': 'blue', 'VIP': 'green'}
shapes = {'Regular': 'o', 'Premium': 's', 'VIP': '^'}
```

```python
# Step 3: Plot Age vs Income (Proximity & Similarity)
plt.figure(figsize=(10,6))
for ctype in data['CustomerType'].unique():
    subset = data[data['CustomerType'] == ctype]
    plt.scatter(subset['Age'], subset['Income'],
                c=colors[ctype],
                marker=shapes[ctype],
                s=100,
                alpha=0.7,
                edgecolors='black',
                label=f'{ctype} Customers')

# Step 4: Add Enclosure for Education groups
for edu in data['Education'].unique():
    subset = data[data['Education'] == edu]
    plt.gca().add_patch(plt.Rectangle(
        (subset['Age'].min()-1, subset['Income'].min()-1000),
        subset['Age'].max()-subset['Age'].min()+2,
        subset['Income'].max()-subset['Income'].min()+2000,
        fill=False, edgecolor='orange', linestyle='--', linewidth=1.5,
        label=f'{edu} Education' if edu=='High School' else ""
    ))

# Step 5: Connection - Visit frequency over customers
plt.plot(data['Age'], data['VisitFrequency']*5000, color='purple', linestyle='-', linewidth=1.5, label='Visit Frequency (scaled)')

# Step 6: Add labels, title, legend
plt.title('Gestalt Principles in Data Visualization')
plt.xlabel('Age')
plt.ylabel('Income')
plt.legend(loc='upper left', bbox_to_anchor=(1,1))
plt.grid(True)
plt.show()
```

```python
# Step 1: Create sample dataset
np.random.seed(404)
fitness_data = pd.DataFrame({
    'Day': np.arange(1, 31),                              # Days of the month
    'Steps': np.random.randint(3000, 15000, 30),          # Daily steps
    'CaloriesBurned': np.random.randint(200, 800, 30),    # Daily calories burned
    'WorkoutType': np.random.choice(['Cardio', 'Strength', 'Yoga'], 30),
    'Mood': np.random.choice(['Happy', 'Neutral', 'Tired'], 30)  # Mood after workout
})
```

```python
workout_colors = {'Cardio': 'red', 'Strength': 'blue', 'Yoga': 'green'}
workout_shapes = {'Cardio': 'o', 'Strength': 's', 'Yoga': '^'}

mood_enclosure = {'Happy': {'color': 'yellow', 'alpha': 0.2},
                  'Neutral': {'color': 'gray', 'alpha': 0.1},
                  'Tired': {'color': 'purple', 'alpha': 0.15}}

plt.figure(figsize=(12, 8))

for workout_type in fitness_data['WorkoutType'].unique():
    subset = fitness_data[fitness_data['WorkoutType'] == workout_type]
    plt.scatter(subset['Steps'], subset['CaloriesBurned'],
                c=[workout_colors[wt] for wt in subset['WorkoutType']],
                marker=workout_shapes[workout_type],
                s=100,
                alpha=0.7,
                edgecolors='black',
                label=f'Workout Type: {workout_type}')

for mood in fitness_data['Mood'].unique():
    subset = fitness_data[fitness_data['Mood'] == mood]
    min_steps, max_steps = subset['Steps'].min(), subset['Steps'].max()
    min_cal, max_cal = subset['CaloriesBurned'].min(), subset['CaloriesBurned'].max()

    plt.gca().add_patch(plt.Rectangle(
        (min_steps - 100, min_cal - 10),
        (max_steps - min_steps) + 200,
        (max_cal - min_cal) + 20,
        fill=True,
        color=mood_enclosure[mood]['color'],
        alpha=mood_enclosure[mood]['alpha'],
        edgecolor='none',
        label=f'Mood: {mood}' if mood == 'Happy' else ""
    ))

ax2 = plt.gca().twinx()
ax2.plot(fitness_data['Day'], fitness_data['Steps'], color='darkorange', linestyle='-', linewidth=2, label='Steps over Day (Connection)')
ax2.set_ylabel('Steps', color='darkorange')
ax2.tick_params(axis='y', labelcolor='darkorange')

plt.title('Gestalt Principles in Fitness Data Visualization')
plt.xlabel('Steps')
plt.ylabel('Calories Burned')

lines, labels = plt.gca().get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
plt.legend(lines + lines2, labels + labels2, loc='upper left', bbox_to_anchor=(1.05, 1))


plt.grid(True)
plt.show()
```

**Data-Ink Ratio**

Preattentive attributes and Gestalt principles are important for understanding which features can be used to visualize data and how the brain processes these visual cues. However, it's easy to overuse these elements, which can reduce their effectiveness in highlighting and distinguishing key information. A key guideline for creating effective data visualizations is that charts and graphs should clarify the data, helping the audience gain insights and understanding. They should avoid excessive clutter that could obscure the data or make it hard to interpret.

A common way to approach this principle is through the concept of maximizing the data-ink ratio. This ratio compares the amount of "data-ink" used to the total ink in a table or chart, where data-ink refers to the ink necessary for conveying the data’s meaning to the audience. Non-data-ink, on the other hand, includes elements that do not contribute to understanding the data. For example, a pie chart might use color and a legend to differentiate between categories, whereas a bar chart can convey the same information without these extra elements, resulting in a higher data-ink ratio.

```python
# Step 1: Create sample dataset
np.random.seed(404)
fitness_data = pd.DataFrame({
    'Day': np.arange(1, 31),                              # Days of the month
    'Steps': np.random.randint(3000, 15000, 30),          # Daily steps
    'CaloriesBurned': np.random.randint(200, 800, 30),    # Daily calories burned
    'WorkoutType': np.random.choice(['Cardio', 'Strength', 'Yoga'], 30),
    'Mood': np.random.choice(['Happy', 'Neutral', 'Tired'], 30)  # Mood after workout
})
```

```python
# Step 2: Visualize the data using a Pie Chart
workout_type_counts = fitness_data['WorkoutType'].value_counts()

plt.figure(figsize=(8, 6))
plt.pie(workout_type_counts, labels=workout_type_counts.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Workout Type Distribution (Pie Chart)')
plt.show()
```

```python
# Step 3: Visualize the data using a Bar Chart
plt.figure(figsize=(8, 6))
plt.bar(fitness_data['Day'], fitness_data['Steps'], color=plt.cm.Paired.colors)
plt.title('Daily Steps (Bar Chart)')
plt.xlabel('Day')
plt.ylabel('Steps')
plt.show()
```

```python
# Step 4: Estimate Data-Ink Ratios (Approximate values)
# For simplicity, we assume the following:
# - Pie chart uses 3 non-data elements (legend, colors, labels)
# - Bar chart uses 2 non-data elements (axes, labels)

pie_chart_data_ink = len(workout_type_counts) / (len(workout_type_counts) + 3)  # Pie chart: 3 non-data ink elements
bar_chart_data_ink = len(fitness_data) / (len(fitness_data) + 2)  # Bar chart: 2 non-data ink elements

# Step 6: Print Data-Ink Ratios for Pie and Bar Charts
print(f"Pie Chart Data-Ink Ratio: {pie_chart_data_ink:.2f}")
print(f"Bar Chart Data-Ink Ratio: {bar_chart_data_ink:.2f}")
```

**Other Data Visualization Design Issues**

 Minimizing Eye Travel <br>

 ata visualizations should be easy to interpret, revealing insights while minimizing cognitive load. This can be achieved by applying preattentive attributes, Gestalt principles, and increasing the data-ink ratio. Reducing eye travel also helps in making charts more efficient.

For example, the Office of Budget and Performance Improvement for Springfield compares the clearance rates of two police districts. Below is a chart comparing the clearance rates for property crimes in Districts 1 and 2 over the last six months.

The chart features that increase eye travel, common in Excel defaults, include the legend placed at the bottom, requiring viewers to move their eyes between the chart and legend. Moving the legend closer to the lines or labeling them directly would reduce this effort. Additionally, rotated axis titles force unnecessary eye movement. Aligning titles within the chart would improve readability by minimizing eye travel.

 Choosing a Font for Text

 Text is a key element in data visualizations, used for labeling axes, filling table values, and highlighting important details. Since text plays such a crucial role, the choice of font is an important design consideration. Most data visualization tools, including Excel, offer a wide variety of font options.

While experts may differ on the ideal font to use, there is general consensus that some fonts are better suited for data visualizations than others. Specifically, sans-serif fonts (which lack the small projections at the end of characters) are typically preferred over serif fonts (which include these projections, called serifs). For example, common serif fonts include Times New Roman and Courier, while popular sans-serif fonts include Arial, Calibri, and Verdana.

Serif fonts are generally more suited for print, while sans-serif fonts are favored for digital displays, as they tend to be more legible at smaller sizes. Since data visualizations are often viewed both in print and digitally and involve varying font sizes, sans-serif fonts are usually the better choice. For instance, the charts in this textbook use Calibri, the default font in Excel, while printed charts use Myriad Pro due to its readability at different sizes. Other sans-serif fonts like Arial and Verdana are also acceptable.

More important than whether the font is serif or sans-serif is the consistency and restraint in font usage. Data visualizations should use the same font type and size for similar information. For example, the titles for horizontal and vertical axes should typically be in the same font and size. Inconsistent font sizes can mislead the viewer by suggesting that certain elements have more importance. Using too many different fonts can clutter the visualization and increase cognitive load. Therefore, it's best to stick to one font type, using size variations, bold, italics, or color to differentiate elements and guide the audience’s attention.

**Common Mistakes in Data Visualization Design**


Wrong Type of Visualization

The choice of chart or table for data visualization largely depends on the audience and the insights or message you want to convey. While this textbook offers best practices for designing effective visualizations, many decisions about which chart or design elements to use will vary depending on the specific context and goal of the visualization. In this section, we explore different situations where one type of visualization might be more appropriate than another. However, it's important to remember that the most effective visualization always depends on the audience's needs and the message being communicated.

When the goal is to present precise numerical values, tables are often a better choice than charts. Bar or column charts are typically preferred over pie charts because they make it easier for the audience to compare values, leveraging the preattentive attribute of length, which is more effective than using shape. Nonetheless, the right choice of visualization always depends on the objective, and it's not always immediately clear which one is best.

 Trying to Display Too Much Information

 A common mistake in creating effective charts is attempting to convey too much information on a single chart, which can overwhelm the audience by presenting too many insights at once.

For example, consider Keeland Industries, an online company that sells both OEM (original-equipment manufacturer) and aftermarket replacement parts for automobiles. Since Keeland sells parts nationwide, it divides the U.S. into 12 regions for tracking sales performance. The management team is primarily interested in comparing OEM sales across these regions to identify top performers and doing the same for aftermarket sales. Trying to combine both comparisons into one chart could lead to confusion, so it's more effective to separate these insights into distinct visualizations.

image.png

 Too Many Attributes

 The "Too Many Attributes" mistake occurs when a chart is overloaded with excessive visual elements, such as multiple colors, shapes, sizes, or labels, making it cluttered and difficult to interpret. This can confuse the audience and increase cognitive load, as viewers struggle to discern the most important insights. To avoid this, focus on a few key attributes and simplify the design, possibly using multiple charts to convey different insights instead of overcrowding one chart with too much information. This ensures the visualization remains clear and easy to understand.

 Unnecessary use of 3D

  Three-dimensional (3D) charts are often difficult for an audience to understand. Many novice analysts will create 3D charts in Excel even when the third dimension does not provide any useful information. Consider the 3D column chart shown in Figure below. This chart compares the billable hours for eight consultants who work for Stanley Consulting Group.
