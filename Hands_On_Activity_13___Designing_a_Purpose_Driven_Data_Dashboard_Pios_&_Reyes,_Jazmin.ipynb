**1. Objectives:**

This activity aims to demonstrate students’ ability to design purposeful and user-centered data dashboards. Students will analyze the dashboard’s purpose, evaluate user requirements, and apply engineering principles such as clarity, accuracy, and interactivity in creating effective visual designs.

**2. Intended Learning Outcomes (ILOs):**

**3. Discussions:**

A well-designed data dashboard is more than a collection of charts — it’s a purposeful storytelling tool that connects data to decisions.

Understanding the Purpose of the Data Dashboard:
Every dashboard must have a clear objective — whether it’s to monitor performance, analyze patterns, or inform strategic decisions. The design should align with this goal, focusing only on the most relevant data and metrics.

Considering the Needs of the Data Dashboard’s Users:
Dashboards should be designed with the user in mind. A manager may need high-level KPIs, while an analyst may need granular, filterable data. Knowing the user’s goals, skills, and workflow ensures that the dashboard supports, rather than overwhelms, their decision-making.

Data Dashboard Engineering:
Engineering a dashboard involves organizing visual elements efficiently, ensuring smooth interactivity, and maintaining data accuracy. It also includes performance considerations (like handling large datasets), responsive design, and consistent color schemes.

Step 1: Install and Import Required Libraries

```python
!pip install plotly pandas
import pandas as pd
import numpy as np
import plotly.express as px
```

Step 2: Create a Sample Dataset

We’ll simulate student performance data to design an academic performance dashboard.

```python
# Create dataset
np.random.seed(10)
students = [f"Student {i}" for i in range(1, 21)]
subjects = ["Math", "Science", "English", "Programming", "Data Analytics"]

data = {
    "Student": np.random.choice(students, 100),
    "Subject": np.random.choice(subjects, 100),
    "Score": np.random.randint(60, 100, 100),
    "Attendance (%)": np.random.randint(70, 100, 100)
}

df = pd.DataFrame(data)
df.head()
```

Step 3: Design Dashboard Components

```python
# Average Score per Subject
fig1 = px.bar(df.groupby("Subject", as_index=False)["Score"].mean(),
              x="Subject", y="Score", title="Average Score per Subject",
              color="Score", color_continuous_scale="Blues")

# Attendance vs. Score Scatter
fig2 = px.scatter(df, x="Attendance (%)", y="Score", color="Subject",
                  title="Attendance vs. Performance", hover_data=["Student"])

# Score Distribution
fig3 = px.box(df, x="Subject", y="Score", title="Score Distribution by Subject",
              color="Subject")

fig1.show()
fig2.show()
fig3.show()
```

Step 4: Combine into One Interactive Dashboard

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

# Aggregated data
avg_scores = df.groupby("Subject", as_index=False)["Score"].mean()
avg_attendance = df.groupby("Subject", as_index=False)["Attendance (%)"].mean()

# Dashboard layout
fig = make_subplots(rows=2, cols=2,
                    subplot_titles=("Average Score per Subject", "Attendance vs. Score",
                                    "Score Distribution", "Average Attendance per Subject"),
                    specs=[[{"type": "bar"}, {"type": "scatter"}],
                           [{"type": "box"}, {"type": "bar"}]])

fig.add_trace(go.Bar(x=avg_scores["Subject"], y=avg_scores["Score"], name="Average Score"), row=1, col=1)
fig.add_trace(go.Scatter(x=df["Attendance (%)"], y=df["Score"], mode="markers", name="Scatter Plot"), row=1, col=2)
fig.add_trace(go.Box(x=df["Subject"], y=df["Score"], name="Score Distribution"), row=2, col=1)
fig.add_trace(go.Bar(x=avg_attendance["Subject"], y=avg_attendance["Attendance (%)"], name="Average Attendance"), row=2, col=2)

fig.update_layout(height=700, width=950, title_text="🎓 Student Performance Dashboard")
fig.show()
```

**Task 2:** Design for User Needs

Modify the dashboard to meet specific user needs:

If your user is a teacher, highlight underperforming students.

If your user is a student, highlight their personal progress vs. class average.

If your user is a department head, summarize subject-level averages only.

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import numpy as np
import pandas as pd

np.random.seed(10)
students = [f"Student {i}" for i in range(1, 21)]
subjects = ["Math", "Science", "English", "Programming", "Data Analytics"]

data = {
    "Student": np.random.choice(students, 100),
    "Subject": np.random.choice(subjects, 100),
    "Score": np.random.randint(60, 100, 100),
    "Attendance (%)": np.random.randint(70, 100, 100)
}

df = pd.DataFrame(data)

avg_scores = df.groupby("Subject", as_index=False)["Score"].mean()
avg_attendance = df.groupby("Subject", as_index=False)["Attendance (%)"].mean()

underperforming_threshold = 70

fig_teacher = make_subplots(rows=2, cols=2,
                    subplot_titles=("Average Score per Subject", "Attendance vs. Score (Highlighting Underperformers)",
                                    "Score Distribution", "Average Attendance per Subject"),
                    specs=[[{"type": "bar"}, {"type": "scatter"}],
                           [{"type": "box"}, {"type": "bar"}]])

fig_teacher.add_trace(go.Bar(x=avg_scores["Subject"], y=avg_scores["Score"], name="Average Score"), row=1, col=1)

fig_teacher.add_trace(go.Scatter(x=df["Attendance (%)"], y=df["Score"], mode="markers", name="Scatter Plot",
                                 marker=dict(color=np.where(df["Score"] < underperforming_threshold, 'red', 'blue'))), row=1, col=2)

fig_teacher.add_trace(go.Box(x=df["Subject"], y=df["Score"], name="Score Distribution"), row=2, col=1)
fig_teacher.add_trace(go.Bar(x=avg_attendance["Subject"], y=avg_attendance["Attendance (%)"], name="Average Attendance"), row=2, col=2)

fig_teacher.update_layout(height=700, width=950, title_text="🎓 Student Performance Dashboard (Teacher View)")
fig_teacher.show()
```

```python
sample_student = "Student 10"

student_df = df[df["Student"] == sample_student]

student_avg_score = student_df["Score"].mean()
student_avg_attendance = student_df["Attendance (%)"].mean()

fig_student = make_subplots(rows=2, cols=2,
                    subplot_titles=(f"{sample_student}'s Scores vs. Class Average", f"{sample_student}'s Attendance vs. Class Average",
                                    f"{sample_student}'s Score Distribution by Subject", "Class Score Distribution by Subject"),
                    specs=[[{"type": "bar"}, {"type": "bar"}],
                           [{"type": "bar"}, {"type": "box"}]])

fig_student.add_trace(go.Bar(x=avg_scores["Subject"], y=avg_scores["Score"], name="Class Average Score"), row=1, col=1)
fig_student.add_trace(go.Bar(x=student_df["Subject"], y=student_df["Score"], name=f"{sample_student}'s Score"), row=1, col=1)

fig_student.add_trace(go.Bar(x=avg_attendance["Subject"], y=avg_attendance["Attendance (%)"], name="Class Average Attendance"), row=1, col=2)
fig_student.add_trace(go.Bar(x=student_df["Subject"], y=student_df["Attendance (%)"], name=f"{sample_student}'s Attendance"), row=1, col=2)

fig_student.add_trace(go.Bar(x=student_df["Subject"], y=student_df["Score"], name=f"{sample_student}'s Scores by Subject"), row=2, col=1)

fig_student.add_trace(go.Box(x=df["Subject"], y=df["Score"], name="Class Score Distribution"), row=2, col=2)


fig_student.update_layout(height=700, width=950, title_text=f"🎓 {sample_student}'s Performance Dashboard")
fig_student.show()
```

**5. Supplementary Activity:**

Design Your Own Dashboard: “Smart Campus Energy Monitor”
Create a dashboard that shows the energy usage of different campus areas (e.g., laboratories, offices, classrooms) over time.

1. Simulate or use sample data (Energy consumption per hour or per day).

2. Include at least three charts (bar, line, pie).

3. Ensure your design reflects a clear purpose and target user (e.g., facilities manager).

4. Write a short explanation of your engineering decisions (why you chose certain colors, layouts, or interactivity).

```python
num_time_points = 30
campus_areas = ['Laboratories', 'Offices', 'Classrooms', 'Library', 'Gym']

dates = pd.date_range(start='2023-01-01', periods=num_time_points, freq='D')

np.random.seed(42)
data = []
for date in dates:
    for area in campus_areas:
        if area == 'Laboratories':
            consumption = np.random.randint(500, 1500)
        elif area == 'Offices':
            consumption = np.random.randint(100, 500)
        elif area == 'Classrooms':
            consumption = np.random.randint(200, 800)
        elif area == 'Library':
            consumption = np.random.randint(150, 600)
        elif area == 'Gym':
            consumption = np.random.randint(300, 1000)
        data.append([date, area, consumption])

energy_df = pd.DataFrame(data, columns=['Date', 'Campus Area', 'Energy Consumption (kWh)'])

display(energy_df.head())
```

```python
import plotly.express as px

total_consumption_over_time = energy_df.groupby("Date")["Energy Consumption (kWh)"].sum().reset_index()
fig_line = px.line(total_consumption_over_time, x="Date", y="Energy Consumption (kWh)",
                   title="Total Energy Consumption Over Time")
fig_line.show()

average_consumption_per_area = energy_df.groupby("Campus Area")["Energy Consumption (kWh)"].mean().reset_index()
fig_bar = px.bar(average_consumption_per_area, x="Campus Area", y="Energy Consumption (kWh)",
                 title="Average Energy Consumption per Campus Area",
                 color="Campus Area")
fig_bar.show()

total_consumption_per_area = energy_df.groupby("Campus Area")["Energy Consumption (kWh)"].sum().reset_index()
fig_pie = px.pie(total_consumption_per_area, values="Energy Consumption (kWh)", names="Campus Area",
                 title="Proportion of Total Energy Consumption per Campus Area")
fig_pie.show()
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

fig_dashboard = make_subplots(rows=2, cols=2,
                              subplot_titles=("Total Energy Consumption Over Time", "Average Energy Consumption per Campus Area",
                                              "Proportion of Total Energy Consumption per Campus Area"),
                              specs=[[{"type": "scatter", "colspan": 2}, None],
                                     [{"type": "bar"}, {"type": "pie"}]])

fig_dashboard.add_trace(fig_line.data[0], row=1, col=1)

fig_dashboard.add_trace(fig_bar.data[0], row=2, col=1)

fig_dashboard.add_trace(fig_pie.data[0], row=2, col=2)

fig_dashboard.update_layout(height=700, width=900, title_text="Smart Campus Energy Monitor")
fig_dashboard.show()
```
